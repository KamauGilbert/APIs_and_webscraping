```python
#import libraries
from randomuser import RandomUser
import pandas as pd
```

```python
r= RandomUser()
some_list= r.generate_users(10)
some_list
```

```python
#getting the full names, email, gender and phone numbers of the users
for user in some_list:
  print(user.get_full_name(), " ", user.get_email(), " ", user.get_gender(), " ", user.get_phone())
```

```python
#let's generate photos of the ten users using the get_picture() fnction
for user in some_list:
  print(user.get_picture())
```

```python
#let's create a pandas dataframe to store all this information
def user_list():
  users=[]
  for user in some_list:
    users.append({'Name':user.get_full_name, 'email':user.get_email(), 'Gender':user.get_gender, 'Photo':user.get_picture})
  return pd.DataFrame(users)


user_list()
```

#**Fruityvice API**

```python
#import libraries
import requests
import json
```

We will proceed to obtain the fruityvice API data using requests.get('url')

```python
data = requests.get('https://fruityvice.com/api/fruit/all')
data
```

```python
results = json.loads(data.text)
results_table=pd.DataFrame(results)
results_table.head()
```

The results_table is in a nested json format. The nutrition column contains multiple subcolumns, so the data needs to be flattened/normalized.

Flattening data is a useful technique for transforming complex, nested data structures into simpler, one-dimensional forms, facilitating easier processing and analysis.

```python
df= pd.json_normalize(results)
df.head()
```

```python
# Extracting the genus of the fruit with the name 'Tomato'

def get_genus():
    name = input('Enter name: ')
    try:
        genus_of_name = df.loc[df['name'] == name, 'genus']
        if genus_of_name.empty:
            print(f"No genus found for the name: {name}")
        else:
            print(genus_of_name.values[0])
    except KeyError:
        print(f"Invalid column name or name not found: {name}")

get_genus()
```
